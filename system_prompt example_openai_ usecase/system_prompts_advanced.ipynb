## Helper Functions

Let's create helper functions to manage conversations with different system prompts.

```python
def chat_with_history(messages: List[Dict[str, str]], temperature: float = 0) -> str:
    """Chat with the model using a message history

    Args:
        messages: List of message dictionaries with role and content
        temperature: Controls randomness in the response (0 to 1)

    Returns:
        Model's response
    """
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature
    )
    return response.choices[0].message.content

def simulate_conversation(system_prompts: Dict[str, str], conversation: List[Dict[str, Any]]) -> None:
    """Simulate a conversation with multiple roles

    Args:
        system_prompts: Dictionary mapping roles to their system prompts
        conversation: List of conversation turns with role and message
    """
    history = []

    for turn in conversation:
        role = turn["role"]
        message = turn["message"]

        # Add system prompt for the current role
        if role in system_prompts:
            history.append({"role": "system", "content": system_prompts[role]})

        # Add user message
        history.append({"role": "user", "content": message})

        # Get model's response
        response = chat_with_history(history)

        # Print the exchange
        print(f"{role}: {message}")
        print(f"Assistant: {response}")
        print("-"*75)

        # Add assistant's response to history
        history.append({"role": "assistant", "content": response})
```
